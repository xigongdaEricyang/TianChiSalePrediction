```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../CleanData/withMax_trainData.csv')
```

```python
def getColumns(pattern):
    return [c for c in df.columns.values if pattern.match(c)]            
classids, years, months = getColumns(re.compile('class_id_')), getColumns(re.compile('year_')), getColumns(re.compile('month_'))
```

```python
gp = df.drop_duplicates([c for c in df.columns.values if c not in years+months+['sale_quantity']]).groupby(classids)
```

```python
for k, g in gp:
    print(k)
```
